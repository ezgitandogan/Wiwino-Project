```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv(r"C:\Users\ezgit\Desktop\Wiwino-Project\CSVs\2.csv")

# Set Seaborn style
sns.set(style="whitegrid")

# Create a vertical bar chart: Country Name vs. Number of Wineries
plt.figure(figsize=(6, 4))
colors = sns.color_palette("Set2", n_colors=len(df))  # Change color palette
bars = plt.bar(df['country_name'], df['wineries_count'], color=colors)

# Add the number of wineries above each bar
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # Position the text in the middle of the bar
        bar.get_height() + 0.5,  # Position the text slightly above the bar
        f'{int(bar.get_height())}',  # Display the number of wineries
        ha='center',
        fontsize=12
    )

# Labeling the chart
plt.xlabel('Country Name', fontsize=14)
plt.ylabel('Number of Wineries', fontsize=14)
plt.title('Number of Wineries by Country', fontsize=16, weight='bold')

# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45, ha='right')

# Adjust layout to make room for labels
plt.tight_layout()

# Show the plot
plt.show()

```

```python

# Read the CSV file
df = pd.read_csv(r"C:\Users\ezgit\Desktop\Wiwino-Project\CSVs\4.csv")

# Count the number of unique wines for each keyword and group
keyword_group_wine_counts = df.groupby(['keywords_name', 'group_name'])['wines_name'].nunique().reset_index()

# Rename columns for clarity
keyword_group_wine_counts.columns = ['Keyword Name', 'Group Name', 'Number of Wines']

# Set Seaborn style
sns.set(style="whitegrid")

# Create a bar chart
plt.figure(figsize=(14, 8))
barplot = sns.barplot(x='Keyword Name', y='Number of Wines', hue='Group Name', data=keyword_group_wine_counts, palette='viridis')

# Add annotations
for p in barplot.patches:
    barplot.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

# Labeling the chart
plt.xlabel('Keyword Name', fontsize=14)
plt.ylabel('Number of Wines', fontsize=14)
plt.title('Number of Wines Associated with Each Keyword by Group', fontsize=16, weight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.legend(title='Group Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()





```

```python
# Assuming you have a 'price' column in your data
df = pd.read_csv(r"C:\Users\ezgit\Desktop\Wiwino-Project\CSVs\7.csv")

# Extract relevant columns
df_filtered = df[['wines_names', 'ratings_average']]

# Remove duplicates to get unique country names with their average ratings
df_unique = df_filtered.drop_duplicates()

# Set Seaborn style
sns.set(style="whitegrid")

# Scatter plot example
plt.figure(figsize=(10, 7))
scatter_plot = sns.scatterplot(
    x='wines_names', 
    y='ratings_average',
    data=df_unique,
    hue='ratings_average',
    palette='viridis',  # Color map for the scatter plot
    s=100,  # Marker size
    edgecolor='w',  # Marker edge color
)

# Add annotations
for i, row in df_unique.iterrows():
    scatter_plot.annotate(f'{row["ratings_average"]:.2f}', 
                          (row["wines_names"], row["ratings_average"]),
                          ha='center', va='center', xytext=(0, 5), textcoords='offset points')

# Labeling the chart
plt.xlabel('Wine Name', fontsize=14)
plt.ylabel('Average Rating', fontsize=14)
plt.title('Top 5 Cabernet Sauvignon Wines by Average Rating', fontsize=16, weight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()



```
